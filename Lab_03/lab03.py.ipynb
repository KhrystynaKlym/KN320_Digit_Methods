##### Згенеруємо декілька масивів різними способами та виконаємо базові математичні операції

```python
l = (3, 3)

O = np.ones(l)
print('Матриця О:\n', O)
R = np.random.random(l)
print('Матриця R:\n', R)
Z = np.zeros(l)
print('Матриця Z:\n', Z)
C = np.arange(l[0])
print('Матриця C:\n', C)

print(f"""
Віднімання:\n {C - (O + 10)}
Множення:\n {O * R}
Ділення:\n {R / C}
Степінь:\n {C ** Z}
Поріняння:\n {R > 0.5}
Ділення на 0:\n {O / Z}
""")
```

```python
A=np.eye(4)
print('Одинична діагональна матриця, створена методом `np.eye`:\n', A)
print('\nПорівняння елементів матриці А на наявність 1:\n', A==1)
B=np.ones((4,4))
print('\nМатриця В, створена методом `np.ones`:\n', B)
print('\nРізниця матриць (B-A):\n', B-A)
```

### Множення матриць

```python
A = np.array([ [1, 2, 3], [2, 2, 2], [3, 3, 3] ])
B = np.array([ [3, 2, 1], [1, 2, 3], [-1, -2, -3] ])
print(f"Маємо два масиви:\n {A} \n\n {B}")

R = A * B
print(f"\nПоелементне множення масивів:\n {R}")

R = np.dot(A, B)
print(f"\nТут, масиви перемножуються як матриці:\n {R}")

MA = np.mat(A)
MB = np.mat(B)

R = MA * MB
print(f"\nТут масиви перетворені на матриці і далі перемножуються:\n {R}")
```

### Вирівнювання розмірностей

```python
A = np.array([ [11, 12, 13], [21, 22, 23], [31, 32, 33] ])
B = np.array([1, 2, 3])
C = B[:, np.newaxis]

print(f"Вектор В є рядком з трьох елементів {B.shape}: {B}, тому додавання (A+B) буде по рядках:\n {A + B}")
print(f"""\nВектор C є стовпцем {C.shape}:
{C}
\nтому додавання буде по стовпцях:
{A + C}""")
print(f"\nУ випадку додавання рядка до стовпця результуючою буде матриця, в якої доповнюються рядки і стовпці:\n {B + C}")
```

### Конкатенація, повторення та зміна форми (shape)

```python
D = np.ones(3)
Z = np.zeros((3,1))
print('Матриця D:', D)
print('Матриця Z:\n', Z)
print(f"""
При конкатенації рядків маємо: {np.concatenate((D, D))}")
\nПри конкатенації стовпців маємо:\n{np.concatenate((Z, Z))}
\nМасив D {D} має розмірність {D.shape} тому може виконати операцію конкатенації з В {B} розмірністю {B.shape}:
{np.concatenate((B, D))}
\nДля конкатенації з С
{C}
де С має розмірність = {C.shape} потрібно змінити розмірність D = {D.shape}
\nЦе можна зробити так:
{D[:, np.newaxis].shape}
або
{D.reshape((3,1)).shape}
в результаті можемо виконати конкатенацію:
{np.concatenate((C, D.reshape((3,1))))}
\nПри повторенні D {D} 2 рази для рядкі та 3 для стовпців будем мати:
{np.tile(D, (2, 3))}
\nЯк бачимо конкатенація та повторення є дещо схожими у деяких випадках:
- конкатенація: {np.concatenate((D, D))}
- повторення: {np.tile(D, 2)}
\nТакож можна перетворити двовимірний масив A розмірністю {A.shape} на рядок: {A.flatten()} з розмірністю {A.flatten().shape}
""")
```

```python
a = np.arange(9).reshape((3, 3))
print('Створений за допомогою `arange` двохвимірний масив `а` розмірністю 3х3:\n', a)
print('\nМасив `а` після конкатенації:\n', np.concatenate((a,a)))
print('\nМасив `а` після повторення:\n', np.tile(a,3))

b = np.arange(10,16).reshape((3, 2))
print('\nСтворений за допомогою `arange` масив `b` розмірністю 3х2:\n', b)
print('\nОскільки, розмірність масивів `а`(3x3) і `b`(3x2) не співпадає, то змінимо розмірність масиву `b`:\n', b.reshape(2,3) ,
'\n\nта виконаємо конкатенацію масивів `a` і `b`:\n', np.concatenate((a, b.reshape(2,3))))
```

```python
print('Створила двовимірний масив `c` (розмірністю 1х12) та знайшла всі можливі комбінації reshape.\n') 
c=np.arange(12).reshape(1,12)
print(c)
c=np.arange(12).reshape(2,6)
print('\n', c)
c=np.arange(12).reshape(3,4)
print('\n', c)
c=np.arange(12).reshape(4,3)
print('\n', c)
c=np.arange(12).reshape(6,2)
print('\n', c)
c=np.arange(12).reshape(12,1)
print('\n', c)
```

### Трансформації масивів

```python
O = np.arange(6).reshape(3,2)
O_r = O.reshape((2,3))
O_f = O.flatten()
O_t = O.transpose()
O_fl = np.flip(O)
O_fa = np.flip(O, axis=1)

print(f"""{O}
Змінити розмірність shape {O.shape} на {O_r.shape} при цьому розмір size {O.size} = {O_r.size} та ndim {np.ndim(O)} = {np.ndim(O_r)}:
{O_r}
Також можна перетворити {np.ndim(O)}-х вимірний масив розмірністю {O.shape} на рядок:
{O_f} 
з розмірністю {O_f.shape} при цьому size = {O_f.size} та ndim стане {np.ndim(O_f)}-х вимірним 
Або транспонувати масив щоб він став {O_t.shape} але зверніть увагу на елементи:
{O_t}
Можна обернути масив (обертаємо по стовпцях і рядках одночасно): 
{O_fl}
Або обернути лише елементи в кожному рядку (або стовпці):
{O_fa}
""")
```

```python
t=np.arange(20,32)
print('Створила вектор `t` з 12 елементів:\n', t, 
'\n\nВизначила:\n`shape` вектора `t`:', t.shape, '\n`size` вектора `t`:', t.size, '\n`ndim` вектора `t`:', t.ndim)
t1=np.arange(20,32).reshape(3,4)
print('\nПеретворила вектор `t` у матрицю `t1` розмірністю 3х4:\n', t1)
t3 = t1.transpose()
t4 = np.flip(t3)
print('\nТранспонувала матрицю `t1`:\n', t3, '\n\nОбернула матрицю `t1` (по стовпцях і рядках одночасно):\n', t4)
t5 = t4.flatten()
print('\nПеретворила отриманий багатовимірний масив назад у вектор (стрічку):\n', t5)
```
