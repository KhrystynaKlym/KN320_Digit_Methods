### Робота з простими лістами

```python
%time
import numpy as np
temperature = [20.1, 20.1, 20.2, 20.4]
print('Список заданих температурних значень: ',temperature)

t1 = temperature * 5
print(f"При множенні списку temperature на 5 отримуємо: {t1}")
```

```python
t1=list()
for x in temperature:
    t1.append(x*5+30)
print(f"При множенні на 5 та збільшенні на 30 значення кожного елементу зі списку temperature отримуємо: {t1}")

t1 = [ x*5+30 for x in temperature] 
print(f"Тепер отримаємо: {t1}")
```

### Типи масивів/векторів, багатовимірні масиви, розуміння вимірів

```python
print('\n!!!HOMEWORK!!!\n')
print("""ndim - кількість вимірів масиву NumPy як ціле число int;
shape - форма (= довжина кожного виміру) масиву NumPy як кортеж;
\tНавіть у випадку одновимірного масиву це кортеж з одним елементом замість цілого значення.
\t!!! Кортеж з одним елементом має кінцеву кому (3,)!!!
\tНаприклад, у випадку двовимірного масиву shape - це (number of rows - кількість рядків, number of columns - кількість стовпців).
\tЯкщо треба отримати лише кількість рядків або стовпців, то можна отримати кожен елемент кортежу.
size - це розмір (= загальна кількість елементів) масиву NumPy""")

import numpy as np
examples = {
    'Скаляр': np.array(7), 
    'Вектор Рядок': np.array([7, 7, 7, 7, 7]), 
    'Вектор Стовпець': np.array([[7], [7], [7]]), 
    'Матриця': np.array([[7, 7],[7, 7], [7, 7], [7, 7], [7, 7]]),
    'Багатовимірна Матриця': np.array([ [[7, 7], [7, 7], [7, 7], [7, 7]],
                                        [[7, 7], [7, 7], [7, 7], [7, 7]],
                                        [[7, 7], [7, 7], [7, 7], [7, 7]],
                                        [[7, 7], [7, 7], [7, 7], [7, 7]] ])
    }
for key in examples:
    print(f"\nРозглядаємо приклад з {key}. \nЗначення x буде: \n", examples[key])
    print(f"Тип {key} x: ", type(examples[key]))
    print(f"Розмірність/вимір {key} x:", np.ndim(examples[key]))
    print(f"Форма {key} x: ", examples[key].shape)
    print(f"Кількість елементів х: {np.size(examples[key])}")
    print("-=-"*10)
```
